# Data Challenge

## Load Packages

```python
import pandas as pd
import numpy as np
```

## Load in Data

```python
allergies = pd.read_feather('../data/allergies.feather')
careplans = pd.read_feather('../data/careplans.feather')
conditions = pd.read_feather('../data/conditions.feather')
encounters = pd.read_feather('../data/encounters.feather')
immunizations = pd.read_feather('../data/immunizations.feather')
medications = pd.read_feather('../data/medications.feather')
observations = pd.read_feather('../data/observations.feather')
patients = pd.read_feather('../data/patients.feather')
procedures = pd.read_feather('../data/procedures.feather')
```

We take a look of what information each dataframe contains one by one.

```python
allergies.head(10)
```

The $\texttt{allergies}$ dataframe contains the information about one allergy description of a patient, with the ID of the patient, ID of the encounter and dates

```python
careplans.head(10)
```

The $\texttt{careplans}$ dataframe contains treatments on a certain reason for an individual patient. The **CODE** column corresponds to treatments (e.g. 53950000 corresponds to 'Respiratory therapy' in the $0^{\text{th}}$ and $6^{\text{th}}$ sample). The **START** and **STOP** columns should be the dates that the patient starts and stops to receive the corresponding treatment.

```python
conditions.head(10)
```

The $\texttt{conditions}$ is a subset of $\texttt{careplans}$, containing the diagnosis of a patient on an encounter.

```python
encounters.head(10)
```

The $\texttt{encounters}$ has the description of an encounter and its reason.

```python
immunizations.head(10)
```

```python
immunizations.loc[0, 'DESCRIPTION']
```

The $\texttt{immunizations}$ contains patients' immunization injection history.

```python
medications.head(10)
```

The $\texttt{medications}$ has the similar structure with $\texttt{careplans}$, which substitutes treatments with medicines.

```python
observations.head(10)
```

The $\texttt{observations}$ contains basic body measure data of patients. 

```python
patients.head(10)
```

The $\texttt{patients}$ contains personal information of patients, such as name, age, ID number, etc.

```python
procedures.head(10)
```

The $\texttt{procedures}$ contains almost the same information as $\texttt{encounters}$, in a different data format.
